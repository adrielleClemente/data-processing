```python
#importação da biblioteca
import pandas as pd
```

```python
#Leitura dos documentos
df1 = pd.read_excel('Aracaju.xlsx')
df2 = pd.read_excel('Fortaleza.xlsx')
df3 = pd.read_excel('Natal.xlsx')
df4 = pd.read_excel('Recife.xlsx')
df5 = pd.read_excel('Salvador.xlsx')
```

```python
#Concatenação dos aruivos
df =pd.concat([df1,df2,df3,df4,df5])
```

## Datas

```python
df['Data']=df['Data'].astype('int64')
```

```python
df.dtypes
```

```python
#Transformar o tipo da coluna Data em tipo date
df['Data']=pd.to_datetime(df['Data'])
```

```python
df.dtypes
```

```python
#Agrupar po ano - dt é datetime, ou seja, ao passar o parâmetro year busca na data o ano  
df.groupby(df['Data'].dt.year)['Receita'].sum()
```

```python
#Criação de uma coluna: Ano
df['Ano_Venda'] = df['Data'].dt.year
```

```python
df.sample(5)
```

```python
#Extração do mês e dia
df['mes_venda'], df['dia_venda']=(df['Data'].dt.month, df['Data'].dt.day)
```

```python
df.sample(5)
```

```python
#Coluna de trimestre
df['trimestre_venda'] = df['Data'].dt.quarter
```

```python
df.sample(5)
```

```python
df.pop('semestre_venda')
```

```python
df.sample(5)
```

```python
#Filtrar vendas, especificamente de um ano e um mês
vendas_mar_19 = df.loc[(df['Data'].dt.year == 2019)&(df['Data'].dt.month == 3)]
```
